# B. DATA CLEANING

**B.1. Handling Missing Values**

```python
# columnwise null values count
bln_df.isnull().sum()
```

*The above info shows the complete count of the null values. We see that the columns 'company.name', 'company.relationship', 'company.sector', 'company.type','demographics.gender', 'wealth.type','wealth.how.category', 'wealth.how.industry' contains null values*

**Checking the percentange of null values in each column is**

```python
for col in bln_df.columns:
    null = bln_df[col].isnull().sum()
    percentage = (null/len(bln_df))*100
    print(col,":", percentage,"%")
```

*Since columns: company.name, company.relationship, company.sector, company.type,demographics.gender,wealth.type,wealth.how.category,wealth.how.industry are having less than 10% null values, droping these NaN values wouldn't affect the dataset much.*

**Dropping missing values in "company.name" Column**


```python
#dropping all the missing values in 'Company.name' column
bln_df.dropna(subset=['company.name'],inplace=True)
```

**Filling missing values in "company.relationship" Column**

```python
#Filling all the missing values of 'company.relationship' column with'not known'
bln_df['company.relationship'].fillna("not known",inplace=True)
bln_df
```

**Filling missing values in "demographic.gender" Column**

```python
#Filling all the missing values in the 'demographic.gender' column with 'not known'
bln_df['demographics.gender'].fillna("not known",inplace=True)
bln_df
```

**Dropping all the remaining missing values**

```python
#Dropping all the remaining values
bln_df.dropna(inplace=True) 
```

```python
# checking whether there is any null value left unhandled
bln_df.isnull().sum()
```

*So our billionaires dataset is now clean i.e. it doesn't contain any null values.*

## **Query 9:  Year wise no of male and female billionaires.**

## **Query 14.Correlation heatmap between the variables**

```python
#droping some columns as correlation is established only between numerical data and then finding their correaltion
corr_df=bln_df.drop(["name","company.name","company.sector"],axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
#Plotting the heatmap using the above information 
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidths=1,linecolor="Black")
heatmap.set_title("correlation heatmap between the variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

#### Correlation heatmap is a type of plot that visualize the strength of relationships between numerical variables. Correlation plots are used to understand which variables are related to each other and the strength of this relationship. A correlation plot typically contains a number of numerical variables, with each variable represented by a column. The rows represent the relationship between each pair of variables. The values in the cells indicate the strength of the relationship, with positive values indicating a positive relationship and negative values indicating a negative relationship. Correlation heatmaps can be used to find potential relationships between variables and to understand the strength of these relationships.
